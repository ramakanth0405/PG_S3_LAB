# Cleaning Scraped Data

06 July 2026, Monday

## Raw Scraped Data

```python
data = "    webscraping using python    " 
```

### White Space Removal

```python
clean_data = data.strip()
clean_data
```

```python
import re
re.sub(r'\s+', ' ', clean_data)
```

### Special Character Removal

```python
humidity = "  70%  "
# Use replace function
clean_humidity = humidity.replace("%", "").strip()
clean_humidity
```

### Converting Data Type

```python
clean_humidity
```

```python
humidity_int = int(clean_humidity)
humidity_int
```

```python
# Checking datatype
type(humidity_int)
```
