Accessing entire contents of a tag using content and children attribute

```python
from bs4 import BeautifulSoup
import requests
filepath = '/Users/ramakanthvprabhu/Desktop/PG_S3/PG_S3_IoT/PG_S3_IoT_0f_web_scrape/IoT_Webscrape_local/PG_S3_IoT_0_sample_web.html'
```

```python
with open(filepath, 'r', encoding='utf-8') as file:
    html_content = file.read()
    soup = BeautifulSoup(html_content, 'html.parser')
```

```python
print(soup.prettify())
```

To get the contents of the body tag in HTML Document

```python
body_content =  soup.body.contents
print(body_content)
```

To print body content without new line characters

```python
for ele in body_content:
    print(ele)
```

```python
for ele in body_content:
    if ele != '\n':
        print(ele)
```

```python
for ele in body_content:
    if ele != '\n':
        print(ele.text)
```

```python
ul_contents = soup.body.ul.contents
print(ul_contents)
```

```python
for ele in ul_contents:
    if ele != '\n':
        print(ele.text)
```
