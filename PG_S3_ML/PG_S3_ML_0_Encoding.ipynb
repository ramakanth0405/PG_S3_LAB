Ordinal Encoding

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
data = {'Grade':['High', 'Medium', 'Low', 'High']}
df = pd.DataFrame(data)
print(df)
```

```python
encoder_ord = OrdinalEncoder(categories = [['Low', 'Medium', 'High']]) # Use [[]]
encoded_data = encoder_ord.fit_transform(df[['Grade']])
print(encoded_data)
```

```python
type(encoded_data)
```

```python
for i in encoded_data:
  if i == [0.]:
    print(i, 'Low')
  elif i ==[1.]:
    print(i, 'Medium')
  else:
    print(i, 'High')
```
