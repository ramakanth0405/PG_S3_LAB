```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
```

We need averaged_perceptron_tagger_eng PoS Model

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
text = ('English texts for beginners to practice reading and comprehension online and for free. Practicing your comprehension of written English will both improve your vocabulary and understanding of grammar and word order. The texts below are designed to help you develop while giving you an instant evaluation of your progress.')
```

```python
tokenized = sent_tokenize(text)
```

```python
tokenized
```

```python
for sentence in tokenized:
    words = word_tokenize(sentence)
    words = [w for w in words if w.lower() not in stop_words]
    tagged = nltk.pos_tag(words)
```

Common PoS (Part of Speech tag)

- CC - Conjunction
- CD - Number
- DT - Determiner
- EX - Existential
- FW - Foreign word
- IN -Preposition
- JJ / JJR / JJS - Adjective / Comparative / Superlative
- NN / NNS / NNP / NNPS - Noun singular/plural / Proper noun
- PRP / PRP$ - Pronoun / Possessive pronoun
- RB / RBR / RBS - Adverb / Comparative / Superlative
- VB / VBD / VBG / VBN / VBP / VBZ - Verb forms
- MD - Modal
- POS - Possessive ending
- RP - Particle
- TO - To
- UH - Interjection
- WDT / WP / WP$ / WRB - Wh-determiners/pronouns/adverb
