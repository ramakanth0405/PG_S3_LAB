```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
text = "This is a sample text for generating n-grams."
```

### strict bigram

```python
vectorizer_22 = CountVectorizer(ngram_range=(2, 2))
```

```python
X = vectorizer_22.fit_transform([text])
```

```python
df = pd.DataFrame(X.toarray(), columns = vectorizer_22.get_feature_names_out())
df
```

### Combination of Unigram and Bigram

```python
vectorizer_12 = CountVectorizer(ngram_range=(1, 2))
```

### Use (𝑚𝑖𝑛_𝑛≤𝑚𝑎𝑥_𝑛)

```python
X_12 = vectorizer_12.fit_transform([text])
```

```python
df_12 = pd.DataFrame(X_12.toarray(), columns = vectorizer_12.get_feature_names_out())
df_12
```
