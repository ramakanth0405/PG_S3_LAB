Linguistic Analysis using regex tokenization and spacy

## Write a python program to perform Regex Based Tokenization using spaCy

```python
import re
import spacy
from spacy.tokenizer import Tokenizer

nlp = spacy.blank("en")

text = "Hello, world! This is a regex-tokenizer example: hello@example.com, josh@example.com"
# Extract email addresses using regex and print only the emails
email_pattern = re.compile(r"[A-Za-z0-9._+-]+@[A-Za-z0-9.-]+\.[A-Za-z]{2,}")
emails = email_pattern.findall(text)
print("Input:\n", text)
print("\nExtracted emails:\n", emails)
```

Extract Date

```python
text_date = 'Hello this is Sam and I was born on 18/12/1996'
date_pattern = re.compile(r"\b\d{2}/\d{2}/\d{4}\b")
dates = date_pattern.findall(text_date)
print("\nExtracted dates (DD-MM-YYYY):\n", dates)
```

Extract Phone Number

```python
# Extract phone numbers (common formats: +CC-123-456-7890, 123-456-7890, (123) 456-7890, 1234567890)
text_phone = 'Hello this is Sam and my phone number is +91 9898989890'
phone_pattern = re.compile(r"\b(?:\+?\d{1,3}[-.\s]?)?(?:\(\d{2,4}\)[-.\s]?|\d{2,4}[-.\s]?)?\d{3,4}[-.\s]?\d{3,4}\b")
phones = phone_pattern.findall(text_phone)
print("\nExtracted phone numbers:\n", phones)
```
