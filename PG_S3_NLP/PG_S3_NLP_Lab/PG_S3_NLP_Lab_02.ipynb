### Write a python program to perform Regex Based Tokenization using spaCy

```python
import re
import spacy
from spacy.tokenizer import Tokenizer

nlp = spacy.blank("en")

text = "Hello, world! This is a regex-tokenizer example: hello@example.com, josh@example.com"
# Extract email addresses using regex and print only the emails
email_pattern = re.compile(r"[\w.+-]+@[\w.-]+\.[A-Za-z]{2,}")
emails = email_pattern.findall(text)
print("Input:\n", text)
print("\nExtracted emails:\n", emails)

# Tokenize using spaCy and print tokens
doc = nlp(text)
tokens = [token.text for token in doc]
print("\nspaCy tokens:\n", tokens)
```

Extract Date

```python
# Extract dates in the format DD/MM/YYYY
text_date = input('Enter text containing dates in the format DD/MM/YYYY: ')
date_pattern = re.compile(r"\b\d{2}/\d{2}/\d{4}\b")
# Use spaCy to tokenize and extract tokens matching the date pattern
doc_date = nlp(text_date)
dates = [token.text for token in doc_date if date_pattern.match(token.text)]
print("\nExtracted dates (DD-MM-YYYY):\n", dates)
```

Extract Phone Number

```python
# Extract phone numbers in the format +91 space followed by 10 digits
text_phone = input("Enter text containing phone numbers in the format +91 1234567890: ")
phone_pattern = re.compile(r"\+91\s\d{10}\b")

# Use spaCy to tokenize and extract tokens matching the phone pattern
doc_phone = nlp(text_phone)
phones = [token.text for token in doc_phone if phone_pattern.match(token.text)]
print("\nExtracted phone numbers:\n", phones)
```

```python
import spacy
import re

# Load a blank English model for minimal overhead
nlp = spacy.blank("en")

# Simplest combined regex pattern for:
# 1. Emails: [\w.-]+@[\w.-]+\.\w+
# 2. Phone (+91 followed by space and 10 digits): \+91 \d{10}
# 3. Dates (DD/MM/YYYY): \d{2}/\d{2}/\d{4}
regex_pattern = re.compile(r"[\w.-]+@[\w.-]+\.\w+|\+91\d{10}|\d{2}/\d{2}/\d{4}")

# Override the default token_match without defining a custom function
nlp.tokenizer.token_match = regex_pattern.match

# Sample text to test extraction
text = "Please reach out to support@example.com or call +919876543210 by 25/12/2026."

# Process the text
doc = nlp(text)

# Extract and print only the tokens that match our specific regex patterns
extracted_tokens = [token.text for token in doc if regex_pattern.match(token.text)]

print("Extracted Tokens:")
for i in extracted_tokens:
    print(f"- {i}")
```
