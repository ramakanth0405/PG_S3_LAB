### Write a Python program to load datasets from NLTK Corpora and Perform Text Cleaning Operations

White space, Lowecase, Special Character

```python
import re
import nltk
from nltk.corpus import gutenberg

nltk.download('gutenberg')
```

```python
gutenberg.fileids()
```

```python
print(gutenberg.raw('whitman-leaves.txt'))
```

```python
fid = 'whitman-leaves.txt'
raw = gutenberg.raw(fid)

# Remove extra whitespace (collapse sequences of whitespace to single space)
cleaned = ' '.join(raw.split())

# Convert to lowercase
cleaned = cleaned.lower()

# Remove special characters (keep a-z, 0-9 and spaces)
cleaned = re.sub(r'[^a-z0-9\s]', '', cleaned)

# Logic Replace all the characters with an empty string except 
# the lower case letters a to z, numbers 0 to 9, and a single space

print(cleaned)
```
