### AIM
Write a python program to perform text normalization using lowercasing, noise removal, stopword removal, stemming, lemmatization, and vectorization using Bag-of-Words and TF-IDF techniques

### Procedure

* Import required libraries
* Define a sample text corpus
* Convert text to lowercase
* Remove special characters using regular expressions
* Remove stopwords using NLTK
* Apply porter and snowball stemming
* Apply WordNet and spaCy lemmatization
* Generate Bag-of-Words representation using CountVectorizer
* Generate TF-IDF representation using TfidfVectorizer
* Display the processed the processed text and feature matrices

```python
import re
import nltk
import spacy
```

```python
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, SnowballStemmer
from nltk.stem import WordNetLemmatizer

from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
#Sample Documents
documents = [
    'Natural language processing is exciting field of Artificial Intelligence.',
    'Text preprocessing improves NLP Model Performance.', 
    'NLP techniques makes the language understanding very easy.'
    'It is very generous and courageous of you to share your knowledge with us.',
]
```

```python
# Lowercase the documents
clean_docs = []
for doc in documents:
    doc = doc.lower()
    doc = re.sub(r'[^a-z\s]', '', doc)

    word = doc.split()

    stop_words = set(stopwords.words('english'))
    word = [w for w in word if w not in stop_words]
    clean_docs.append(' '.join(word))
```

```python
print("Cleaned Documents:")
print(clean_docs)
```

```python
# Stemming
porter_stemmer = PorterStemmer()
snowball_stemmer = SnowballStemmer('english')
```

```python
# Porter Stemmer
for doc in clean_docs: 
    print([porter_stemmer.stem(word) for word in doc.split()])
    
```

```python
for doc in clean_docs:
    print([snowball_stemmer.stem(word) for word in doc.split()])
```
