## Part of Speech (PoS) Tagging 

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
```

```python
nltk.download('averaged_perceptron_tagger_eng')
```

```python
stop_words = set(stopwords.words('english'))
text = input('Enter text for Pos Tagging: ')
```

Sentence Tokenization

```python
tokenized = sent_tokenize(text)
print(tokenized)
```

Word Tokenization

```python
for i in tokenized:
    words = word_tokenize(i)
    words = [w for w in words if w.lower() not in stop_words]
    tagged = nltk.pos_tag(words)
```

Common PoS (Part of Speech tag)

- CC - Conjunction
- CD - Number
- DT - Determiner
- EX - Existential
- FW - Foreign word
- IN -Preposition
- JJ / JJR / JJS - Adjective / Comparative / Superlative
- NN / NNS / NNP / NNPS - Noun singular/plural / Proper noun
- PRP / PRP$ - Pronoun / Possessive pronoun
- RB / RBR / RBS - Adverb / Comparative / Superlative
- VB / VBD / VBG / VBN / VBP / VBZ - Verb forms
- MD - Modal
- POS - Possessive ending
- RP - Particle
- TO - To
- UH - Interjection
- WDT / WP / WP$ / WRB - Wh-determiners/pronouns/adverb
