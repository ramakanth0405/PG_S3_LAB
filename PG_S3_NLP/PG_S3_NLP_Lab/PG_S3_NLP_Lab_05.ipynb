# Next word prediction using ngrams

### AIM 
### To predict the next word in a sentence using the trigram language model

### Procedure

1.  Import required libraries
2.  Input a sample text corpus 
3.  Convert the text into lowecase
4.  Tokenize the text into words
5.  Generate Trigrams
6.  Store Trigrams in a dictionary
7.  Accept two words from the user 
8.  Predict the most likely next word


```python
from collections import defaultdict, Counter
import nltk
from nltk.tokenize import word_tokenize
from nltk.util import trigrams
import random
```

nltk.download("punkt_tab", quiet=True)

```python
corpus = (
    "The cat chased the mouse. The cat caught the mouse. "
    "The dog chased the cat. The cat slept on the mat."
)

words = word_tokenize(corpus.lower())
```

```python
model = defaultdict(list)
for w1, w2, w3 in trigrams(words):
    model[w1, w2].append(w3)
```

```python
model
```

```python
user_input = input("Enter two words separated by a space: ")
input_words = user_input.lower().split()
if len(input_words) == 2:
    history = tuple(input_words)

    if history in model:
        possible_words = model[history]
        
        most_likely_word = max(possible_words, key=possible_words.count)
        
        print(f"The most likely next word (highest frequency) is: '{most_likely_word}'")
    else:
        print("Word combination not found in the corpus.")
else:
    print("Please enter exactly two words.")
```
