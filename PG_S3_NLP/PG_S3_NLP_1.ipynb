# Write a Python program to remove stopwords using NLTK

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
nltk.download('punkt_tab')
```

```python
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
text_1 = input('Enter Text to be Word Tokenized: ')
stop_words = set(stopwords.words('english'))
tokens = word_tokenize(text_1.lower())
```

```python
print(tokens)
```

Filtering tokens to remove stopwords

```python
filter_tokens = [word for word in tokens if word not in stop_words]
print(filter_tokens)
```
