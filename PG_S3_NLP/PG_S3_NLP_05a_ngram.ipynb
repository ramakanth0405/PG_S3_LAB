````python
import pandas as pd
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk import ngrams
````

````python
text = "This is a sample text for generating n-grams. I love NLP and n-grams are very useful in text analysis."
````

## Performing Word Tokenization

````python
text_1 = word_tokenize(text)
````

````python
text_1
````

### Using ```ngrams``` function to generate bigrams

````python
gen_ngrams = ngrams(text_1, 2)
````
