### Unsupervised pre-trained machine learning model for NLTK to split text into sentences or individual words

Word Tokenize

```python
from nltk.tokenize import word_tokenize
text_1 = input('Enter Text to be Word Tokenized: ')
words = word_tokenize(text_1)
print(words)

# ['Hello', 'this', 'is', 'Ramakanth']
```

Sentence Tokenize

```python
from nltk.tokenize import sent_tokenize
text_2 = input('Enter Sentence to be Sentence Tokenized')
sentences = sent_tokenize(text_2)
print(sentences)
```
