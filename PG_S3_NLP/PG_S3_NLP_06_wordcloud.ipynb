```python
from collections import Counter
```

```python
text = 'Big fish has fins. Big fish can swim. Big fish lives in the water'
```

```python
from nltk.tokenize import word_tokenize
```

```python
text_clean = word_tokenize(text.lower())
```

```python
text_clean
```

```python
frequency = Counter(text_clean)
```

```python
frequency
```

```python
import seaborn as sns
```

Check x and y values

```python
sns.barplot(x=list(frequency.values()), y=list(frequency.keys()), orient = 'h')
plt.title('Word Frequency Visualization')
plt.xlabel('Words')
plt.ylabel('Frequency')
```

```python
from wordcloud import WordCloud
```

```python
wordcloud = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(frequency)
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud)
plt.axis('off')
plt.title("IMDB Movie Reviews Word Cloud")
plt.show()
```
