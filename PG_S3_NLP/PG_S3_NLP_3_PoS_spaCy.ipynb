```python
nlp = spacy.load('en_core_web_sm')
```

```python
text = ("\
I enjoy writing articles on GeeksforGeeks checkout\
my other article by going to my profile section.")
```

```python
doc = nlp(text)
```

```python
for token in doc:
    print(token, token.pos_)
```
